```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# Connection parameters
username = "root"
password = "Alexx123."
host = "localhost"
port = 3306  # must be integer
database = "us_project"

# Encode password (safe for special characters)
password_encoded = quote_plus(password)

# Create SQLAlchemy engine
engine = create_engine(
    f"mysql+pymysql://{username}:{password_encoded}@{host}:{port}/{database}",
    connect_args={"local_infile": 1}  # optional for CSV imports
)

print("Connection engine created successfully!")
```

```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\Usuario\Desktop\Analyst Builder\3 MySQL for Data Analytics\11 Full Projects\4 US Household Income Data Cleaning\USHouseholdIncome_Statistics.csv"
    
)
print(df.head())
```
