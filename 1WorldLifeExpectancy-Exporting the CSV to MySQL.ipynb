```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# Connection parameters
username = "root"
password = "Alexx123."
host = "localhost"
port = 3306  # must be integer
database = "world_life_expectancy"

# Encode password (safe for special characters)
password_encoded = quote_plus(password)

# Create SQLAlchemy engine
engine = create_engine(
    f"mysql+pymysql://{username}:{password_encoded}@{host}:{port}/{database}",
    connect_args={"local_infile": 1}  # optional for CSV imports
)

print("Connection engine created successfully!")
```

```python
# Create df and paste the file path for the CSV

import pandas as pd

df = pd.read_csv(r"C:\Users\Usuario\Desktop\Analyst Builder\3 MySQL for Data Analytics\11 Full Projects\2 World Life Expectancy Data Cleaning/WorldLifeExpectancy.csv")

print(df.head())
```

```python
#First do the following:
#1 Create a database in MySQL: 
#CREATE DATABASE world_life_expectancy;
#2 Verify the database exists
#SHOW DATABASES;

# Is giving me a leading/trailing spaces error (1166, "Incorrect column name 'Life expectancy '") so lets clean columns names
# Clean column names
df.columns = (
    df.columns
    .str.strip()           # remove leading/trailing spaces
    .str.replace(" ", "_") # replace spaces with underscores
)

# Export to sql
df.to_sql(
    name="world_life_expectancy",   # Table name
    con=engine,
    if_exists="replace",              # or "append" to add without deleting existing
    index=False,                      # Avoid exporting pandas index
    chunksize=1000                    # Handle large datasets in batches
)

```
